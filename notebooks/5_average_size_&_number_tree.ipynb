## Average size of trees by district

```python
mean_by_arr = df.groupby("arrondissement")['hauteur_m'].mean().sort_values(ascending=False)
```

```python
mean_by_arr.plot(kind='bar', title='average size of trees by district', ylabel='Average size (m)',
         xlabel='District (arrondissement)', figsize=(6, 5));
```

It can be seen that the average of the trees changes little from one district to another.

## number of Trees by district in paris

```python
df_arrondissement_group = df.groupby("arrondissement")['id'].count()
df_arrondissement_group.sort_values(ascending=False, inplace=True)
```

```python
df_arrondissement_group.plot(kind='bar', title='Trees numbers by district in paris', ylabel='number',
         xlabel='District', figsize=(6, 5));
```

We can see the center of Paris (1 to 11 district) contains fewer trees than the most remote districts
