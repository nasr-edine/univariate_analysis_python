# Average size of trees by district

```python
mean_by_arr = df.groupby("arrondissement")['hauteur_m'].mean().sort_values(ascending=False)
```

```python
mean_by_arr.plot(kind='bar', title='average size of trees by district', ylabel='Average size (m)',
         xlabel='District (arrondissement)', figsize=(6, 5))
```
